## Package mpmath

```python
import mpmath as mp
# Setting precision
mp.mp.dps = 100
```

```python
M_mp=mp.matrix([[0.8165598, 0.6241732, 0.6240969, 0.6240302],
[0.8597795, 0.6958361, 0.6957673, 0.6957071],
[0.8597966, 0.6958662, 0.6957974, 0.6957372],
[0.8598116, 0.6958925, 0.6958237, 0.6957635]])
```

```python
invM_mp=M_mp**-1
```

```python
product=invM_mp * M_mp
```

#### Precision element de la diagonale

```python
for i in range(4):
    print(product[i,i])
```

#### Precision des autres elements

```python
for i in range(4):
    for j in range(4):
        if i != j:
            print(product[i,j])
```

### Application de la methode a la sous matrice

```python
sM = mp.matrix([[0.8165598, 0.6241732, 0.6240969],
[0.8597795, 0.6958361, 0.6957673],
[0.8597966, 0.6958662, 0.6957974]])
sM_inv = sM ** -1
```

```python
product2=sM_inv * sM
```

```python
for i in range(3):
    print(product2[i,i])
```

```python
for i in range(3):
    for j in range(3):
        if i != j:
            print(product[i,j])
```
