## I - Monotonous Strain (Increasing or Decreasing)

### I. Influence of the Strain slope (Positive, High value slopes)

```python
from Functions import RMA_newton
```

```python
N = 300 #Strain Size 

E = 200000
b = 0.001
A = 1000
a = 0.2

strains_final=[0.01,0.05,0.1,0.15,0.2] #Final value of strain, starting value always zero

strains=[]
stress=[]

for obj in strains_final:
    new_strain=(obj/N)*np.array(range(N+1))
    strains.append(new_strain)
    rma=RMA_newton(new_strain)
    rma.activate(E,A,b,a)
    stress.append(np.array(rma.stress))
```

```python
X=np.array(range(301))

plt.rcParams["figure.figsize"] = (18,6)
plt.subplot(1,3,1) #Darker color
plt.plot(X,strains[0],color="#0873ec",lw=2)  
plt.plot(X,strains[1],color="#ffce00",lw=2)  
plt.plot(X,strains[2],color="#07d800",lw=2)  
plt.plot(X,strains[3],color="#f92008",lw=2)  
plt.fill_between(X, np.zeros(301), strains[0],color='#2784ec',alpha=.2)
plt.fill_between(X, strains[0], strains[1],color='#ffdb00',alpha=.2)
plt.fill_between(X, strains[1], strains[2],color='#08e700',alpha=.2)
plt.fill_between(X, strains[2], strains[3],color='#f96251',alpha=.2)
plt.ylabel(r'$\epsilon$(t)',size=12)
plt.xlabel(r'$t$',size=12)
plt.title('Strain with respect to Time',size=14)

plt.subplot(1,3,2)
plt.plot(X,stress[0],color="#0873ec",lw=2)  
plt.plot(X,stress[1],color="#ffce00",lw=2)  
plt.plot(X,stress[2],color="#07d800",lw=2)  
plt.plot(X,stress[3],color="#f92008",lw=2)  
plt.fill_between(X, np.zeros(301), stress[0],color='#2784ec',alpha=.2)
plt.fill_between(X, stress[0], stress[1],color='#ffdb00',alpha=.2)
plt.fill_between(X, stress[1], stress[2],color='#08e700',alpha=.2)
plt.fill_between(X, stress[2], stress[3],color='#f96251',alpha=.2)
plt.ylabel(r'$\sigma$(t)',size=12)
plt.xlabel(r'$t$',size=12)
plt.title('Stress with respect to Time',size=14)

plt.subplot(1,3,3)
plt.plot(strains[3],stress[3],color="#f92008",lw=2)
plt.plot(strains[2],stress[2],color="#07d800",lw=2)  
plt.plot(strains[1],stress[1],color="#ffce00",lw=2)
plt.plot(strains[0],stress[0],color="#0873ec",lw=2)  
plt.ylabel(r'$\sigma$(t)',size=12)
plt.xlabel(r'$\epsilon$(t)',size=12)
plt.title('Stress vs. Strain',size=14)
plt.show()
```

The slope of the strain just indicates how quickly we perform the deformation on steel. The slope of the stress adjusts with respect to this idea. Indeed, for example, the strain sample with highest slope has an almost vertical slope for stress and in contrast the one with smallest strain slope has a significantly smaller stress slope at first. Note also that the stress always has somekind of linear component at first (elastic deformation ?). 

Finally, the strain-stress diagram shows overlapping curves. Slope of strain indication of how far we go along the general curve.

## II - Oscillations in the Strain (Alternating between increasing and decreasing strain)

### I. Back to Zero Linear Strain (Positive Setting)

```python
import random
```

```python
N = 300 #Strain Size 

E = 200000
b = 0.001
A = 1000
a = 0.2

max_strains=[0.001,0.005,0.0075,0.01]
strains=[]
stress=[]

stopping=int(random.uniform(1,N-1))

for obj in max_strains:
    init_slope = obj/stopping
    new_slope  = (init_slope*stopping)/(stopping-N)
    bias       = -new_slope*N
    arr1       = init_slope*np.array(range(stopping+1))
    arr2       = new_slope*np.array(range(stopping+1,N+1))+bias
    new_strain = np.concatenate([arr1,arr2])  
    strains.append(new_strain)
    rma=RMA_newton(new_strain)
    rma.activate(E,A,b,a)
    stress.append(np.array(rma.stress))
```

```python
X=np.array(range(301))

plt.rcParams["figure.figsize"] = (18,6)
plt.subplot(1,3,1) #Darker color
plt.plot(X,strains[0],color="#0873ec",lw=2)  
plt.plot(X,strains[1],color="#ffce00",lw=2)  
plt.plot(X,strains[2],color="#07d800",lw=2)  
plt.plot(X,strains[3],color="#f92008",lw=2)
plt.axvline(stopping,color='k',alpha=0.2)
plt.ylabel(r'$\epsilon$(t)',size=12)
plt.xlabel(r'$t$',size=12)
plt.title('Strain with respect to Time',size=14)

plt.subplot(1,3,2)
plt.plot(X,stress[0],color="#0873ec",lw=2)  
plt.plot(X,stress[1],color="#ffce00",lw=2)  
plt.plot(X,stress[2],color="#07d800",lw=2)  
plt.plot(X,stress[3],color="#f92008",lw=2) 
plt.axvline(stopping,color='k',alpha=0.2)
plt.ylabel(r'$\sigma$(t)',size=12)
plt.xlabel(r'$t$',size=12)
plt.title('Stress with respect to Time',size=14)

plt.subplot(1,3,3)
plt.plot(strains[3],stress[3],color="#f92008",lw=2)
plt.plot(strains[2],stress[2],color="#07d800",lw=2)  
plt.plot(strains[1],stress[1],color="#ffce00",lw=2)
plt.plot(strains[0],stress[0],color="#0873ec",lw=2)  
plt.ylabel(r'$\sigma$(t)',size=12)
plt.xlabel(r'$\epsilon$(t)',size=12)
plt.title('Stress vs. Strain',size=14)
plt.show()
```

Note the elastic behavior of the blue curve !

```python
X=np.array(range(301))

plt.rcParams["figure.figsize"] = (18,6)
plt.subplot(1,3,1) #Darker color
plt.plot(X,strains[0],color="#0873ec",lw=2)  
plt.plot(X,strains[1],color="#ffce00",lw=2)  
plt.plot(X,strains[2],color="#07d800",lw=2)  
plt.plot(X,strains[3],color="#f92008",lw=2)
plt.axvline(stopping,color='k',alpha=0.2)
plt.ylabel(r'$\epsilon$(t)',size=12)
plt.xlabel(r'$t$',size=12)
plt.title('Strain with respect to Time',size=14)

plt.subplot(1,3,2)
plt.plot(X,stress[0],color="#0873ec",lw=2)  
plt.plot(X,stress[1],color="#ffce00",lw=2)  
plt.plot(X,stress[2],color="#07d800",lw=2)  
plt.plot(X,stress[3],color="#f92008",lw=2) 
plt.axvline(stopping,color='k',alpha=0.2)
plt.ylabel(r'$\sigma$(t)',size=12)
plt.xlabel(r'$t$',size=12)
plt.title('Stress with respect to Time',size=14)

plt.subplot(1,3,3)
plt.plot(strains[3],stress[3],color="#f92008",lw=2)
plt.plot(strains[2],stress[2],color="#07d800",lw=2)  
plt.plot(strains[1],stress[1],color="#ffce00",lw=2)
plt.plot(strains[0],stress[0],color="#0873ec",lw=2)  
plt.ylabel(r'$\sigma$(t)',size=12)
plt.xlabel(r'$\epsilon$(t)',size=12)
plt.title('Stress vs. Strain',size=14)
plt.show()
```

* Oscillaton in the deformation $\rightarrow$ oscillation in the constraint
* The more frequent the oscillation in the strain $\rightarrow$ the higher the value of the stress

Everytime we deform a material we need to apply more constraint to deform it again. The exponential decay in the absolute value of the strain characterized this idea. Indeed, even though we decrease the deformation we keep increasing the constraint.
