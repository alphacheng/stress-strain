```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.arima_process as ap
import nbimporter
from Functions import RMA_newton
```

#### The idea here would be to consider that the strain can be modeled by an ARMA process

```python
ar_coef = [0.05]
ma_coef = [10.5]
nobs = 300
y = ap.arma_generate_sample(ar_coef, ma_coef, nobs)
#y2 = ap.arma_generate_sample(ar_coef, ma_coef, nobs)
strain=0.01*y/max(abs(y))
```

```python
plt.plot(strain)
plt.xlabel('t')
plt.ylabel('$\epsilon(t)$')
plt.show()
```

```python
E = 200000
b = 0.001
A = 1000
a = 0.2

rma=RMA_newton(strain)
rma.activate(E,A,b,a)
```
